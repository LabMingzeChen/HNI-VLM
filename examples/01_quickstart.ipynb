## 3. Classify one image

```python
from hni_vlm import HNIClassifier

model = HNIClassifier(backend='qwen')

# Use any URL, local path, PIL image, or bytes
img_url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/2/2f/Stanley_Park_Vancouver.jpg/1280px-Stanley_Park_Vancouver.jpg'

result = model.predict(img_url)

print('Primary category        :', result.primary_category)
print('Social context          :', result.social_context)
print('Activity intensity      :', result.activity_intensity)
print('Photographer engagement :', result.photographer_engagement)
```

## 4. Process a small batch

```python
import pandas as pd

urls = [
    'https://example.com/setting_example.jpg',  # replace with real URLs
    'https://example.com/subject_example.jpg',
    'https://example.com/partner_example.jpg',
]

results = model.predict_batch(urls)

df = pd.DataFrame([r.to_dict() for r in results])
df
```
